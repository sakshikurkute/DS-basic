```python
# write a python program to print deviation of upper letter and lower letter 
# in list format from the given string by using for loop
```

```python
s1 = "SaKshIKurkuTE"
```

```python
lower_letters = []
upper_letters = []

for i in s1 :
    if i.isupper()==True :
        upper_letters.append(i)
    else :
        lower_letters.append(i)
print(upper_letters)
print(lower_letters)
```
